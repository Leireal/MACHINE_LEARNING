```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

```

```python
df= pd.read_csv('../data/processed/datos_normalizados.csv', delimiter=',')

df
```

# BALANCEO DE LOS DATOS POSIBLE TARGET

```python
class_counts = df['label2'].value_counts()
class_counts
```

```python
plt.figure(figsize=[30,10])
sns.heatmap (df.corr(numeric_only=True), annot=True, cmap='coolwarm', vmin=-1);
```

```python
cols_a_borrar = ['setting1', 'setting2', 'setting3', 's1','s5','s6','s10','s16', 's18', 's19', 'failure_cycle','is_failure_cycle']
df_red = df.drop(columns=cols_a_borrar)
df_red
```

```python
df_red['cumcount'] = df_red.groupby('id').cumcount()
df_filtered = df_red[df_red['cumcount'] >= df_red['cumcount'].max() * 0.2]
df_filtered = df_filtered.drop(columns='cumcount')
df_filtered
```

```python
class_counts = df_filtered['label2'].value_counts()
class_counts
```

```python
plt.figure(figsize=[30,10])
sns.heatmap (df_filtered.corr(numeric_only=True), annot=True, cmap='coolwarm', vmin=-1);
```

```python
X = df_filtered.drop(columns=['id', 'cycle', 'label1', 'label2', 'RUL'])
y = df_filtered['label2']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

print(X_train.shape)
print(y_train.shape)
```

```python
# Normalización (muy importante para SVM)
scaler = StandardScaler().fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', decision_function_shape='ovr', random_state=42)
svm.fit(X_train_s, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', 0.01, 0.1, 1],
    'kernel': ['rbf', 'linear']
}

grid = GridSearchCV(SVC(decision_function_shape='ovr', random_state=42),
                    param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid.fit(X_train_s, y_train)

print("Mejores parámetros:", grid.best_params_)
print("Mejor precisión CV:", grid.best_score_)
best_svm = grid.best_estimator_
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_svm.predict(X_test_s)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

CON LOS MEJORES PARAMETROS

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Separar características y variable objetivo
#X = df[['s2','s3','s4','s7','s8','s9','s11','s12','s13','s14','s15','s17','s20','s21']]
#y = df['label2']

# División del dataset con estratificación
#X_train, X_test, y_train, y_test = train_test_split(
 #   X, y, test_size=0.2, stratify=y, random_state=42
#)

# Normalización
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelo con los mejores parámetros encontrados
svm_model = SVC(C=0.1, kernel='rbf', gamma='scale', decision_function_shape='ovr')
svm_model.fit(X_train_scaled, y_train)
```

```python
# Predicciones y evaluación
y_pred = svm_model.predict(X_test_scaled)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.calibration import CalibrationDisplay

disp = CalibrationDisplay.from_estimator(
    svm_model, X_test, y_test, n_bins=10)
plt.title("Curva de calibración de SVM")
plt.show()
```
